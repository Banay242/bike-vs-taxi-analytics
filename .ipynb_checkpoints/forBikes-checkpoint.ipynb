Импорт необходимых библиотек для работы с файлами данных

```python
import pandas as pd
from sqlalchemy import create_engine
```

Грузим данные (важно указать правильные пути к файлам)

```python
df = pd.read_csv('datasets/NYC-BikeShare-2015-2017-combined.csv', index_col=0)
```

```python
df.head()
```

```python
df.dropna()
df = df.drop_duplicates()
```

```python
df
```

Фильтрация по разумной длительности поездки

```python
#df = df[(df['Trip_Duration_in_min'] >= 1) & (df['Trip_Duration_in_min'] <= 180)]
```

```python
df.columns
```

```python
df['Start Time'] = pd.to_datetime(df['Start Time'])
df['Stop Time'] = pd.to_datetime(df['Stop Time'])
df = df[df['Start Time'] - df['Stop Time'] < pd.Timedelta(minutes=180)]
```

```python
df['Birth Year'] = df['Birth Year'].fillna(0).astype(int)
```

```python
df['User Type'] = df['User Type'].astype(str)
```

```python
df['Start Station Name'] = df['Start Station Name'].astype(str)
```

Переименуем для удобства использования 

```python
df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]
df.columns
```
